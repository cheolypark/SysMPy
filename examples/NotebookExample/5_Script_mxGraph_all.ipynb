```python

p = Process("process")

p_con1 = p.Condition("Condition 1")

for i in range(10):
    p2 = p_con1.Process(f"P{i}")

    p_con2 = p2.Condition(f"Condition {i}_2")
    p21 = p_con2.Process(f"P{i}_1")
    p22 = p_con2.Process(f"P{i}_2")  
    
    for j in range(20):
        p21.Action(f"Action1_{i}_{j}") 
        p22.Action(f"Action2_{i}_{j}") 
    
show(p)
```
